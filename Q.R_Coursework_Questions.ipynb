### Question 6

Using the `starwars` dataset, identify the individuals with:

- the tallest height
- the lowest mass

HINT: the starwars dataset is part of the dplyr package

```r
library(tidyverse)
data(starwars)

# Find the individual with the tallest height
tallest <- starwars %>%
  filter(height == max(height, na.rm = TRUE)) %>% # Filter the dataset to calculate the max height, ignoring NA values 
  select(name, height) # Selects the name and height to be displayed after fidning the tallest individual
tallest

# Find the individual with the lowest mass
lightest <- starwars %>%
  filter(mass == min(mass, na.rm = TRUE)) %>% # Filter the dataset to calculate the min mass, ignoring NA values 
  select(name, mass) 
lightest
```

### Question 7

The `dplyr` dataset `storms` is part of the US National Oceanic and Atmospheric Administration database of Atlantic hurricanes. It contains multiple repeated measurements from hurricanes. For a given storm, there are measurements from every 6 hours across the life time of the storm.

Use a tidyverse approach with `dplyr` to answer the following:

- how many different storms are contained in the database?
- which year had the most storms?
- which storm, in which year had the highest pressure?

HINT:
unique() will help ensure you remove duplicated data from a vector e.g.
x <- c(1, 1, 2, 2, 3, 3, 4, 4)
unique(x)

```r
library(dplyr)
data(storms)

# Count the number of different storms
nstorms <- storms %>%
  summarise(amount = n_distinct(name)) #Counts the number of unique names on the dataset
unique(nstorms) # prints the function while removing duplicated data

# Identify year with the most storms
moststorms <- storms %>%
  group_by(year) %>% # groups the data by the year column
  summarise(count = n_distinct(name)) %>%
  arrange(desc(count)) # Orders the data in descending order, highest at the top
unique(moststorms[1, ]) # prints the first row storm, proving the highest

# Identfiy the storm and year with the highest pressure
highpress <- storms %>%
  filter(pressure == max(pressure, na.rm = TRUE)) %>%
  select(name, year, pressure)
unique(highpress) # removes duplicated data
```

### Question 8

The `ggplot2` dataset `diamonds` contains the price and information from the sale of 54,000 different diamonds. You can use ?diamonds to find out more about the dataset if you want.

Using `ggplot2`:

- plot the relationship between the carat (i.e. the weight of the diamond) and the price.
- plot a histogram of prices of all the diamonds
- plot a boxplot of the difference in price between the different cuts

```r
# load the diamonds dataset
library(ggplot2)
data(diamonds)
# ?diamonds used to get extra information about the dataset to add to the graphs

# plotting Carat and Price relationship
ggplot(diamonds, aes(carat, price)) + # plot graph and add the x and y aesthetics
   geom_point( alpha = 0.5, colour = "red") + # make points partially transparent and adding colour 
   labs( title = "Scatterplot of Carat vs Price", 
        x = "Carat", y = "Price ($USD)") # labelling the axis and title

# histogram for diamonds prices
ggplot(diamonds, aes(price)) +
   geom_histogram (binwidth = 500, fill = "orange") + # adding colour and adjusting the width of each bar
   labs( title = "Diamond Price Distribution", 
        x = "Price ($USD)", y = "Count")

# boxplot of prices by the cut
ggplot(diamonds, aes(cut, price)) +
   geom_boxplot( fill = "yellow", colour = "black") +
   labs( title = "Price Differences between Dimaond Cuts", 
        x = "Cut", y = "Price ($USD)")
```

### Question 10

Sticking with the `diamonds` dataset for this question. A carat is a special unit of weight for gemstones where a single carat is 200 mg or 0.2 g. Write a function to convert the `diamonds$carat` vector into grams or milligrams. How heavy is the heaviest diamond?

```r
# function to convert the carat to milligrams
conv <- function(carat){
    grams <- carat * 200
    paste(grams, "mg") # multiply the carat value by 200 and add the "mg" unit
}

# adding a new conversion column to the dataset 
diamonds$weight <- conv(diamonds$carat) 
head(diamonds)

# printing the heaviest diamond
heaviest <- diamonds %>%
   filter(weight == max(weight, na.rm = TRUE)) %>% # selects the heaviest diamond based on the the weight vector
   select(carat, weight)
unique(heaviest) # removes duplicated data
          
```
